# as both the tables have one common column we can combine both the tables using merge 
and while using merge ,
how= left gives more importance to df table
Also on = gives the column name which is same in both the table

```python
pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df= pd.merge(df,df_country,on='Country Code',how='left') #in order to combine tables
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
## observation 
when rating is between 4.5 to 4.9 .....excellent
when rating is between 4.0 to 4.4.......very good
when rating is between 3.5 to 3.9 ......good
when rating is between 2.5 to 2.9...average
when rating is between 2.0 to 2.4....poor
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color', data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color', data=ratings, palette=['white','red','orange','yellow','green','green'])
```

observation
1. Not rated count is very high
2. Average rating is between 2.8 to 3.7
As average rating is the above mentioned range, missing values can be assumed as per average.
